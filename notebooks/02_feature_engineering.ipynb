```python
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/ev-range-model/data/UDDS.csv")
df.columns = ["time","speed"]
df.head()
```

```python
df["speed"] = df["speed"]*1.6 # Miles to kmph conversion
```

```python
# Compute acceleration (still using km/h → we’ll convert internally)
dt = 1  # 1s sampling
df['a'] = df['speed'].diff() / dt
df['a'].fillna(0, inplace=True)

# Feature extraction
df_feat = pd.DataFrame({
    'avg_speed': [df['speed'].mean()],
    'max_speed': [df['speed'].max()],
    'std_speed': [df['speed'].std()],
    'percent_idle': [(df['speed'] < 1).mean()],
    'avg_accel': [df['a'][df['a']>0].mean()],
    'avg_decel': [abs(df['a'][df['a']<0].mean())],
    'accel_std': [df['a'].std()],
    'peak_accel': [df['a'].max()],
    'cycle_duration': [len(df)]
})

df_feat
```
